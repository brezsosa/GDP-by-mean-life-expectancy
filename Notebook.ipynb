```python
plt.figure()
plt.scatter(x, y, alpha = 0.75, marker = 'o')
plt.suptitle('Mean life expectancy vs Mean GDP \nacross the 2000 - 2015 period', color = 'Black', y = 1.1)
plt.title('This scatter plot shows how the mean life expectancy varies\n in relation to variations of the mean GDP by country over a 15 year period. \nEach point in the scatter plot represents a unique country of the world.', color = 'Grey', style = 'italic', fontsize = 10)
plt.xlabel('Mean GDP', color = 'Grey')
plt.ylabel('Mean Life Expectancy', color = 'Grey')
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.xticks([])
plt.yticks([])
plt.show()
```
